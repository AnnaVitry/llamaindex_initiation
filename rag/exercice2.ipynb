## Exercice 2 : Créer un RAG avec LlamaIndex avec des IAs externes

```python
# 0) Importer les modules
from dotenv import load_dotenv
import os
from llama_index.embeddings.azure_openai import AzureOpenAIEmbedding
from llama_index.llms.azure_openai import AzureOpenAI
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
```

```python
# 1) Charger les variables d'environnements
load_dotenv()

AZURE_MODEL_NAME = os.getenv("AZURE_MODEL_NAME", "gpt-4.1") # Le nom de votre déploiement LLM
AZURE_EMBEDDING_MODEL_NAME = os.getenv("AZURE_EMBEDDING_MODEL_NAME", "text-embedding-ada-002") # Le nom de votre déploiement d'embedding
```

```python
# 2) Définir le modèle LLM
llm = AzureOpenAI(
    engine=AZURE_MODEL_NAME, 
    model=AZURE_MODEL_NAME,
    temperature=0.1,
    # Passage explicite de la clé
    api_key=os.getenv("AZURE_OPENAI_API_KEY"))
```

```python
# 3) Définir le modèle d'embedding
embed_model = AzureOpenAIEmbedding(
    model=AZURE_EMBEDDING_MODEL_NAME, 
    deployment_name=AZURE_EMBEDDING_MODEL_NAME,
    # Passage explicite de la clé
    api_key=os.getenv("AZURE_OPENAI_API_KEY") 
)
```

```python
# 4) Effectuer les réglages globaux llamaindex avec `Settings`
Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 1024
```
